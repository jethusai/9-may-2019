# problem 3:
### problem statement
plot a line graph on income vs year for the sate california

```python
import pandas as pd
df=pd.read_csv('DataFiles/income.csv')
y = df.values[3]
x = df.columns
x
y
```

```python
years=[]
for i in range(2,len(x)):
    years.append(x[i])
#years
income=[]
for i in range(2,len(y)):
    income.append(y[i])
#income
plt.xlabel('years')
plt.ylabel('income')

plt.plot(years,income)
plt.show()
```

```python
width =1/2
plt.width=1/2
plt.title('income of california')
plt.bar(years,income,width,color='Red')
plt.savefig('DataFiles/jethya.png')
plt.show()
```
